```python
import numpy as np
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```

```python
siny = y.copy()
cosy = np.cos(x)
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlim(-1, 11) #指定x 轴的区间
plt.ylim(-2, 2)   #指定y 轴的区间
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!")
#plt.axis([-1, 11, -2, 2])  指定x,y轴也可以这样写,这样写也是ok的
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```
